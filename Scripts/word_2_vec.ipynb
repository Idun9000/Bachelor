```python
import gensim
import gensim.downloader
import pandas as pd
from gensim.models import KeyedVectors
import numpy as np
```

```python
word2vec = gensim.downloader.load('word2vec-google-news-300')
word2vec.most_similar("cat")
```

```python
word2vec_twit = gensim.downloader.load('glove-twitter-25')
word2vec_twit.most_similar("cat")
```

```python
data = pd.read_csv('/Users/tildeidunsloth/Desktop/bachelor/Bachelor/Data/75kidsubset.csv')

data['item_definition']
item_list = data['item_definition'].tolist()
print(item_list)
len(item_list)

# New list with unique items
unique_items = set(item_list)
len(unique_items)

#Checking type
type(unique_items)

#Type was set so I change it to list using sorted()
unique_items_list = sorted(unique_items)
type(unique_items_list)
```

```python
# With word2vec model
semantic_distances = np.zeros((len(unique_items_list), len(unique_items_list)))

for i in range(len(unique_items_list)):
    for j in range(len(unique_items_list)):
        if unique_items_list[i] in word2vec and unique_items_list[j] in word2vec:
            semantic_distances[i, j] = word2vec.similarity(unique_items_list[i], unique_items_list[j])


print(semantic_distances)
```

```python
# Create a DataFrame from the semantic distance matrix to look for errors/make sure that the output is correct
semantic_distances_df = pd.DataFrame(semantic_distances, columns=unique_items_list, index=unique_items_list)

# Print the DataFrame
print(semantic_distances_df)

semantic_distances_df.to_csv('/Users/tildeidunsloth/Desktop/bachelor/Bachelor/Data/semantic_distances.csv')
```
